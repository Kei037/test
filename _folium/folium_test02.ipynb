```python
import pandas as pd
import folium
import json
```

```python
df = pd.read_csv('./input/소상공인시장진흥공단_상가(상권)정보_경북_202312.csv')
df.head()
```

```python
# 필요한 정보만 데이터 프레임으로 변경 (시군구명, 상권업종중분류명, 위도, 경도)
df1 = df[['표준산업분류명', '시군구명', '위도', '경도', '상호명']]
df1.head()
```

```python
df2 = df1[df1['표준산업분류명'] == '커피 전문점']
df2.head()
```

```python
df3 = df2[df2['시군구명'] == '경산시']
df3.head()
```

```python
df3.info()
```

```python
df.dtypes
```

```python
# 경산시 : 35.875275° N 위도, 128.817535° E 경도
gyeongsan_map = folium.Map(location=[35.875275, 128.817535], zoom_start=11, tiles='Cartodb Positron')
gyeongsan_map
```

```python
# popup: 마우스 클릭시 나옴. tooltip: 마우스 오버시 나옴
# Tooltip() permanent=True : 고정
for name, lat, lng in zip(df3['상호명'], df3['위도'], df3['경도']):
    tooltip = folium.Tooltip(name, permanent=True)
    folium.Marker([lat, lng], tooltip=tooltip).add_to(gyeongsan_map)
gyeongsan_map
```

```python
gyeongsan_map.save('./output/gyeongsan_map.html')
```
