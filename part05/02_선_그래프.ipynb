# 02. 선 그래프

```python
# 에어코리아에서 제공하는 2001년 ~ 2019년 까지의 미세먼지 데이터 'fine_dust.xlsx'를 이용하여 선 그래포 표현.

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1) 2019년 데이터 읽기
data = pd.read_excel('./input/fine_dust.xlsx', index_col='area')
data.head()
```

```python
data2019 = data[2019]
data2019.head()
```

```python
# 2) 2019년 지역별 미세먼지 선 그래프
plt.figure(figsize=(15 ,4))  # 그래프 사이즈
plt.plot(data2019, color='b', marker='o')  # 원형 마커 지정
plt.title('2019 Fine Dust Line Graph')
plt.xlabel('area')
plt.ylabel('micrometer')
plt.grid()  # 격자 표시
plt.show()
```

```python
# 3) 2016 ~ 2019년 미세먼지 선 그래프
# 2016년 데이터를 가져와 chartdata에 저장하고 마커 및 이름을 지정하는 과정을 2019년 까지 반복'
plt.figure(figsize=(15 ,4))
for year in range(2016, 2019):
    plt.plot(data[year], marker='o', label=year)
plt.title('2016 ~ 2019 Fine Dust Line Graph')
plt.xlabel('area')
plt.ylabel('micrometer')
plt.legend()
plt.grid()
plt.show()
```

```python
# 4) 대구 년도별 미세먼지 선 그래프
daegu = data.loc['Daegu']
daegu
```

```python
plt.figure(figsize=(15 ,4))
plt.plot(daegu, color='b', marker='o')
plt.title('Daegu 2001 ~ 2019 Fine Dust Line Graph')
plt.xlabel('year')
plt.ylabel('micrometer')
plt.grid()
plt.show()
```
