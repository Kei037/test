```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../NumPy/input/health_screenings_2020.csv')
data.info()
```

```python
data['gender'].value_counts()
```

```python
man_data = data.loc[data.gender == 1.0, ['gender', 'height']]
woman_data = data.loc[data.gender == 2.0, ['gender', 'height']]
data['gender'].value_counts()
```

```python
plt.figure(figsize=(10,6))
plt.hist(man_data['height'], bins=20,alpha=0.5, label='man')
plt.title('2020 Health Screenings Man Height Histogram')
plt.xlabel('height')
plt.ylabel('frequency')
plt.xlim(120, 200)
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(man_data['height'], bins=20,alpha=0.5 ,label='Man')
plt.hist(woman_data['height'], bins=20,alpha=0.5 ,label='Woman')
plt.title('2020 Health Screenings Man & woman Height Histogram')
plt.xlabel('height')
plt.ylabel('frequency')
plt.xlim(120, 200)
plt.legend()
plt.grid()
plt.show()
```
