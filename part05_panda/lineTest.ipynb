```python
# 에이코리아에서 제공하는 2001년 2019년 까지의 미세먼지 데이터 'fine_dust.xlsx'를 이용하여 선  그래프로 표현

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1) 2019년 데이터 찾기

data = pd.read_excel('../NumPy/input/fine_dust.xlsx', index_col='area')
data.head()
```

```python
data2019 = data[2019]
data2019.head()
```

```python
# 2) 2019년 지역별 미세먼지 선 그래프
plt.figure(figsize=(15, 4)) # x축과 Y 축의 사이즈를 지정해서 글씨가 안겹치게 만듬
plt.plot(data2019, color='b', marker='o')  # 원형 마커 지정
plt.title('2019 Fine Dust Line Graph')
plt.xlabel('area')
plt.ylabel('micrometor')
plt.grid() # 격자 표시  (그래프에 격자 표시 를 추가하여 각 결과값이 어디에 해당하는지 보기 더 편하게 만듬)
plt.show()
```

```python
# 3) 2016 ~ 2019년 미세먼지 선 그래프
# 2016년 데이터를 가져와 chartdata에 지정하고 마커및 이름을 저장하는 과정을 2019까지 반복
plt.figure(figsize=(15, 4))
for year in range(2016, 2020):  #2016년에 시작해서 2019년에 끝나야되서 2020을 적음
    plt.plot(data[year], marker='o', label=year)
plt.title('2016 ~2019 Fine Dust Line Graph')
plt.xlabel('area')
plt.ylabel('micrometor')
plt.grid()
plt.legend()
plt.show()
```
