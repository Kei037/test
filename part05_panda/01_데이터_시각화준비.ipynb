```python
# 그래프를 표현하다 보면 그래프 제목, X축, Y축 이름 등의 그래프에 대한 추가 옵션을 지정할떄가 많음
# 1) 제목

# 그래프 제목은 title() 메서드를 사용하여 표현
# 제목은 기본적으로 영어 사용하는데, 현금을 사용하려면 별도의 설정을 해야함.
```

```python
# 제목 지정
import matplotlib.pyplot as plt

plt.title('Line Graph')
xdate = [2, 4, 6, 8]
ydate = [1, 3, 5, 7]
plt.plot(xdate, ydate)
plt.show()
```

```python
# 2) 범례
# 그래프 범례는 두개 이상으 ㅣ데이터를 표현할떄 사용
# 범례 주기는 plot()을 활용하여 표현하는데 plot()의 label 속성값에 원하는문자열을 작성하고
# 그래프를 그리기전에 legend() 메서드를 실행
# 범례? 아마도 그래프안에서 선들의 이름을 보여주는 장치를 뜻하는것같음
# 범례 지정
plt.title('Legend')
data1 = [2, 4, 6, 8]
data2 = [8, 6, 4, 2]
plt.plot(data1, label='asc')
plt.plot(data2, label='desc')
plt.legend()
plt.show()
```

```python
# 3) 색상
# 그래프 색항은 plot() 메서드에 color 속성을 추가하여 표현

# 색상지정
xdate = [2, 4, 6, 8]
ydate = [1, 3, 5, 7]
plt.plot(xdate, ydate, color='red')
plt.show()
```

```python
# 4) x축 y축 이름저장
xdate = [2, 4, 6, 8]
ydate = [1, 3, 5, 7]
plt.plot(xdate, ydate)
plt.xlabel('X value')
plt.ylabel('Y value')
plt.show()
```

```python
# 5) 그래프 선 모양
# 그래프 선 모양은 plot()의 linestyle 속성을 이용하여 표현.
# linestyle 속성으로 실선은 '-', 파선은 '--', 점쇄선 '-.', 점선은 ':' 기호로 지정

# 그래프 선 모양지정
data1 = [2, 4, 6, 8]
data2 = [8, 6, 4, 2]
plt.plot(data1, color='r', label='dashed', linestyle='--')
plt.plot(data2, color='g ', label='dotted', linestyle=':')
plt.legend()
plt.show()
```

```python
# 6) 그룹 범위 지정
# plot()를 사용하여 그래프를 표현하다 보면 몇몇 점들은 그림의 범위 경계선에 있어서 잘보이지 않을떄가 있음
# xlim() 메서드와 ylim()메서드를 사용하여 그림의 범위를 수동으로 지정할수있음
# xlim()은 x축의 최솟값과 최대값을 지정하고, ylim()은 y축의 최솟값과 최대값을 지정
# plot()의 marker 속성을 지정하면 선에 점 모양을 다양하게 변경할수있음

# 그림 범위 지정
plt.title('X, Y range')
xdate = [10, 20, 30, 40]
ydate = [1, 3, 5, 7]
plt.plot(xdate, ydate, color='b', linestyle='--', marker='o', markersize='10')
plt.xlim(10, 30)
# plt.xlim(0, 50)  # 그래프의 확대 축소 가능 원래 범위보다 넓게 잡으면 (plt.xlim(0, 50) plt.ylim(-5, 15)) 축소  원래 범위보다 작게 잡으면 (plt.xlim(10, 30) plt.ylim(1, 4))확대
# plt.ylim(-5, 15)
plt.ylim(1, 4)
plt.show()
```

```python
# 7) 내장 시각화 옵션
# 판다스를 이용하면 시각화 라이브러리를 임포트 하지 않아도 내장 그래프 도구를 이용하여 기본적인 시각화를 할 수 있음.
# plot()의 kind 속성을 지정하면 다양한 종류의 그래프를 표현할 수 있음.

# 그래프 속성 값
# 옵션 : 종류
# line : 선 그래프
# bar : 막대 그래프 - 수직
# barh : 막대 그래프 - 수평
# his : 히스토그램 그래프
# box : 박스 그래프
# kde : 커널 밀도 그래프
# area : 면적 그래프
# pie : 원형 그래프
# scatter : 산점도 그래프
# hexbin : 고밀도 산점도 그래프

import pandas as pd

my_score = [[60, 90, 95], [80, 75, 100], [65, 85, 90], [85, 70, 90], [95, 90, 85], [75, 85, 90], [85, 80, 75]]
subject = ['kor', 'math', 'eng']
df = pd.DataFrame(my_score, columns=subject)
df
```

```python
df.plot(kind='line')
plt.show()  # 범례를 자동으로 설정했는데 여기서 범례는 테이블 컬럼 명으로 자동 설정됨. 라인의 색상또한 자동으로 생성되었슴
```
