## 3. 북대구 영업소의 2023년 12월 일자별 총 교통량 변화를 선 그래프로

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./input/TCS_대구경북권톨게이트진출입일교통량_1일_분기_2023_4qt.csv', encoding='cp949')
df.head()
```

```python
df1 = df[df['영업소명'] == '북대구']
df1.head()
```

```python
df1.info()
```

```python
df1['집계일자2'] = pd.to_datetime(df1['집계일자'])
df1.info()
```

```python
# 12월만 출력
df2 = df1[df1['집계일자2'].dt.month == 12]
df2.tail()
```

```python
matplotlib.rcParams['font.family'] = 'AppleGothic' # 맑은 고딕으로 지정
matplotlib.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(15, 5))
sns.lineplot(data=df2, x='집계일자2', y='총교통량', errorbar=None)
plt.show()
```
