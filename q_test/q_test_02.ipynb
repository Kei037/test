```python
# 2. 공공데이터 포털에서 '소상공인시장진흥공단_상가(상권)정보'를 검색해 파일데이터 다운로드를 받으시오.
# 압축을 푼 후 대구 관련 파일을 선택해서 작업을 진행하시오. (24년 3월 11일 기준으로 113,640건의 데이터가 있습니다.)

# 1) '상권업종소분류코드'를 이용해서 카페의 데이터만 추출하여 엑셀로 저장하시오. (파일명: 대구_카페.xlsx)
# 2) 구별 카페수를 막대그래프로 출력
# 3) 대구시 전체의 구별 카페수를 파이그래프로 출력
# 4) 수성구는 카페수를 법정동명을 기준으로 동별로 막대그래프로 출력
# 5) 수성구전체의 카페수를 법정동면 기준으로 파이 그래프로 출력해서 대구시의 상권 분석을 하세요
```

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./input/소상공인시장진흥공단_상가(상권)정보_대구_202312.csv')
df.head()
```

```python
# 1) '상권업종소분류코드'를 이용해서 카페의 데이터만 추출하여 엑셀로 저장하시오. (파일명: 대구_카페.xlsx)
df1 = df[df['상권업종소분류명'] == '카페']
df1.head()
```

```python
df2 = df1[df1['상권업종소분류코드'] == 'I21201']
df2.head()
```

```python
df2.to_excel('./output/대구_카페.xlsx')
```

```python
# 2) 구별 카페수를 막대그래프로 출력
df3 = df2.groupby('시군구명', as_index=False).agg(count=('시군구명', 'count'))
df3
```

```python
matplotlib.rcParams['font.family'] = 'AppleGothic' # 맑은 고딕으로 지정
matplotlib.rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(17, 6))

plt.bar(df3['시군구명'], df3['count'])
plt.title('구별 카페수')
plt.xlabel('구')
plt.ylabel('카페수')
plt.show()
```

```python
# 3) 대구시 전체의 구별 카페수를 파이그래프로 출력
plt.figure(figsize=(17, 6))

plt.pie(df3['count'], labels=df3['시군구명'], autopct='%1.1f%%')
plt.title('구별 카페수')
plt.show()
```

```python
# 4) 수성구는 카페수를 법정동명을 기준으로 동별로 막대그래프로 출력
df4 = df2[df2['시군구명'] == '수성구']
df4.head()
```

```python
df5 = df4.groupby('법정동명', as_index=False).agg(count=('법정동명', 'count'))
df5.head()
```

```python
plt.figure(figsize=(17, 6))

plt.bar(df5['법정동명'], df5['count'])
plt.title('동별 카페수')
plt.xlabel('동')
plt.ylabel('카페수')
plt.show()
```

```python
# 5) 수성구전체의 카페수를 법정동면 기준으로 파이 그래프로 출력해서 대구시의 상권 분석을 하세요
plt.figure(figsize=(17, 6))

plt.pie(df5['count'], labels=df5['법정동명'], autopct='%1.1f%%')
plt.title('동별 카페수')
plt.show()
```
