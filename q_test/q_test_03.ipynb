```python
# 3. 2에서 다운받은 데이터를 folium라이브러리를 사용해서 지도에 위치를 표시합니다.
# 1)  이디야 커피 체인점의 데이터만 추출해서 엑셀로 저장하시오. (파일명 : 대구_카페_이디야.xlsx)
# 2) folium라이브러리를 사용해 지도에  마커를 표시하시오. 마커에 마우스 이벤트 없이도 상호명이 나오도록 하시오. 
# 3) 해당 지도 객체를  html 파일로 저장하시오. (파일명 : 대구_카페_이디야_지도.html)
```

```python
import pandas as pd
import folium
```

```python
df = pd.read_excel('./output/대구_카페.xlsx')
df.head()
```

```python
df1 = df[df['상호명'].str.contains('이디야')]
df1.head()
```

```python
df1.to_excel('./output/대구_카페_이디야.xlsx')
```

```python
daegu_map = folium.Map(location=[35.8683, 128.5988], zoom_start=11, tiles='Cartodb Positron')
daegu_map
```

```python
for name, lat, lng in zip(df1['상호명'], df1['위도'], df1['경도']):
    tooltip = folium.Tooltip(name, permanent=True)
    folium.Marker([lat, lng], tooltip=tooltip).add_to(daegu_map)
daegu_map
```

```python
daegu_map.save('./output/대구_카페_이디야_지도.html')
```
